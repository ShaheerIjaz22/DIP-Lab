```python
image = cv2.imread('image.jpeg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
_, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('OFF')
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.axis('OFF')
plt.imshow(binary_image, cmap='gray')
plt.title("Binarized Image")

plt.show()
```

## Task # 3

```python
image = cv2.imread('image.jpeg')
channels = image.shape[2]

if channels == 3:
    print("Image is three-channel image.")
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    _, binarized_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)
    plt.subplot(1, 2, 1)
    plt.axis('OFF')
    plt.imshow(binarized_image, cmap='gray')
    plt.title("Binarized Image")
    
elif channels == 1:
    print("Image is a one-channel image.")
    cv2.imshow('Grayscale Image', image)
    plt.subplot(1, 2, 1)
    plt.axis('OFF')
    plt.imshow(image)
    plt.title("Grayscale Image")
```

## Task # 4

```python
def binarize(image, threshold):
    binarized_image = np.zeros((image.shape[0], image.shape[1]), dtype=np.uint8)
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            if image[i, j] > threshold:
                binarized_image[i, j] = 255
            else:
                binarized_image[i, j] = 0
    return binarized_image

image = cv2.imread('image.jpeg')
gray_image = np.zeros((image.shape[0], image.shape[1]), dtype=np.uint8)

for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        b, g, r = image[i, j]
        gray_intensity = int(0.299 * r + 0.587 * g + 0.114 * b)
        gray_image[i, j] = gray_intensity

binarized_image = binarize(gray_image, 128)
plt.subplot(1, 2, 1)
plt.axis('OFF')
plt.imshow(binarized_image)
plt.title("Binarized Image")
plt.show()
```

## Task # 5

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    cv2.imshow('frame', gray)
    if cv2.waitKey(1) == ord('q'):
        break
 
cap.release()
cv2.destroyAllWindows()
```
