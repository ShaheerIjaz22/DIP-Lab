```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def match_template(image, temp):
    img = image
    template = temp
    w, h = template.shape[0], template.shape[1]
    res = cv2.matchTemplate(img, template, cv2.TM_CCOEFF)
    _, _, _, max_loc = cv2.minMaxLoc(res)  # Get the best match location
    top_left = max_loc
    bottom_right = (top_left[0] + w, top_left[1] + h)
    cv2.rectangle(img, top_left, bottom_right, 0, 2)
    return img
```

```python

image = cv2.imread("img.jpg")
onga = image.copy()
template = cv2.imread("match.jpg")

result_image = match_template(image,template)

plt.figure(figsize=(8,6))
plt.subplot(131)
plt.imshow(onga)
plt.title('Original Image')
plt.xticks([])
plt.yticks([])

plt.subplot(132)
plt.imshow(template)
plt.title('Template')
plt.xticks([])
plt.yticks([])

plt.subplot(133)
plt.imshow(result_image)
plt.title('Matching Result')
plt.xticks([])
plt.yticks([])

plt.show()
```

```python
new_image = cv2.imread("test.jpg")
copy = new_image.copy()
new_template = cv2.imread("tost.jpg")

new_det = match_template(new_image, new_template)

plt.figure(figsize=(8,6))
plt.subplot(131)
plt.imshow(copy)
plt.title('Original Image')
plt.axis('off')

plt.subplot(132)
plt.imshow(new_template)
plt.title('Template')
plt.axis('off')

plt.subplot(133)
plt.imshow(new_det)
plt.title('Matching Result')
plt.axis('off')

plt.show()
```
