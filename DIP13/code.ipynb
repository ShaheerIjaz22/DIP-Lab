```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

##### Task # 1

```python
img = cv2.imread('./face.jpg')
rotated_img = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)

orb = cv2.ORB_create()

kp1, des1 = orb.detectAndCompute(img, None)
kp2, des2 = orb.detectAndCompute(rotated_img, None)

img_with_kp = cv2.drawKeypoints(img, kp1, None, color=(0, 255, 0), flags=0)
rotated_img_with_kp = cv2.drawKeypoints(rotated_img, kp2, None, color=(0, 255, 0), flags=0)

plt.subplot(221)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(222)
plt.imshow(rotated_img, cmap='gray')
plt.title('Rotated Image')
plt.axis('off')

plt.subplot(223)
plt.imshow(img_with_kp, cmap='gray')
plt.title('Image with Keypoints')
plt.axis('off')

plt.subplot(224)
plt.imshow(rotated_img_with_kp, cmap='gray')
plt.title('Rotated Image with Keypoints')
plt.axis('off')
```

```python
handwriting_img = cv2.imread('hand.jpeg', 0)

fast = cv2.FastFeatureDetector_create()

kp = fast.detect(handwriting_img, None)

img_with_kp = cv2.drawKeypoints(handwriting_img, kp, None, color=(255, 0, 0))


plt.subplot(121)
plt.imshow(handwriting_img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(122)
plt.imshow(img_with_kp, cmap='gray')
plt.title('Image with Keypoints')
plt.axis('off')
```
